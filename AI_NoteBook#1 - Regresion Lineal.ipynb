```python
from sklearn.datasets import load_boston
import numpy as np
import matplotlib.pyplot as plt
```

```python
boston = load_boston() # Dataset de Boston
print(boston.DESCR) # Descripcion del dataset
```

Formula minimizar el error cuadr√°tico medio (MCO): $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
X = np.array(boston.data[:, 5]) # Numero medio de habitaciones
Y = np.array(boston.target) # Valor medio del precio

plt.scatter(X, Y, alpha=0.3) # Grafica de los datos

# A√±adimos una columna de 1's para termino independiente
X = np.array([np.ones(506), X]).T
print(X.T)

B = np.linalg.inv(X.T @ X) @ X.T @ Y # Formula de Minimos Cuadrados Ordinarios

plt.plot([4, 9], [B[0] + B[1] * 4, B[0] + B[1] * 9], 'r') # A√±adimos una linea de regresion
plt.show()
```
